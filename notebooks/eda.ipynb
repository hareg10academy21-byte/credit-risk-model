```python
num_cols = df.select_dtypes(include='number').columns

num_cols
```

```python
for col in num_cols:
    df[col].hist(figsize=(6,4))
    plt.title(col)
    plt.show()
```

```python
cat_cols = df.select_dtypes(include='object').columns

cat_cols
```

```python
for col in cat_cols:
    print(df[col].value_counts())
```

```python
for col in cat_cols:
    df[col].value_counts().head(10).plot(kind='bar')
    plt.title(col)
    plt.show()
```

```python
corr = df.corr(numeric_only=True)

corr
```

```python
import seaborn as sns

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True)
plt.show()
```

```python
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(col)
    plt.show()
```

# Key Insights

1. Transaction amounts are highly right-skewed.

2. Several numerical features contain extreme outliers.

3. Some categorical variables are dominated by a few categories.

4. Certain numerical variables show strong correlations.
